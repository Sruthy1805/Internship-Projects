Converting an Image to a pencil Sketch




Adding Libraries

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

Loading and Plotting the orignal Image

```python
img = cv2.imread("SM.jpg.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Sruthy's Image")
plt.show()
```

Converting to Gray Scale

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

Inverting The Image

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

Smoothening The Image

```python
img_smoothening =cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothening,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

Showing The Pencil Sketch

```python
final = cv2.divide(img_gray, 255 - img_smoothening, scale=215)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Sketched Image")
plt.show()
```
